### Statistics :
- Statistics is a set of mathematical methods and tools that enable us to answer important questions about data. It is divided into two categories
###### 1. Descriptive Statistics: 
- Method of organizing data(table, graph are used for it),summerizing data(descriptive values like mean values, avg. score are used for it)and presenting data in informative way is known as descriptive statistics.
###### 2. Inferential Statistics: 
- It is a method used to determine somethings about a population on the basis of a sample.
                  
### Basic terms :

###### 1. Population :
- Whole Dataset.

###### 2. Sample :
- Subset of population.

###### 3. Variable :
- Character/Number/Quantity of each data/element of a population/sample that can measured/counted example - age, sex, income.

###### 4. Data Singular :
- Single value of any variable of population/sample. It can be Number, word or symbal. example - yes, no, 30, 25, ENGG.

###### 5. Data Plural   : 
- All the values of any variable of population/sample. It can be Number, word or symbal. example - yes/no, 30/35/25, DOC/ENGG/Sports_Person.

###### 6. Experiment :
- Planed activity whose result yield a set of data.

###### 7. Parameter :
- A Numeric value that summerizing data for entire population or summary of population.

###### 8. Statistics :
- A Numeric value that summerizing data for sample or summary of sample.

### Random/Unbiased/Probability Sampling

```python
dataset = pd.read_csv("tips.csv")
dataset
```

```python
# simple random sampling
SRS = dataset.sample(100)
print("Simple Random Sample")
SRS
```

```python
# systametic random sampling
SyRS = dataset.iloc[::10,:]
print("Systametic Random Sample")
SyRS
```

```python
dataset.day.unique()
```

```python
# stratified random sampling
StRS = dataset.groupby('day', group_keys=False).apply(lambda x: x.sample(3))  # selecting 3 sample from each strata(3 sample from all unique day)
print("Systametic Random Sample")
StRS
```

```python
# cluster random sampling
CRS = dataset.sample(10).sort_values(by='time') # selecting 10 sample from whole strata(day)
print("Cluster Random Sample")
CRS
```

### Non Random/biased/Non Probability Sampling

source : https://www.scribbr.com/methodology/sampling-methods/

```python
# convenience non random sampling
CNRS = dataset.iloc[40:160,2:5]
print("Convenience Non Random Sample")
CNRS
```

```python
# voluntary non random sampling
VNRS = np.where(dataset['tip'] > 2)
print("Voluntary Non Random Sample")
VNRS
```

```python
# judgement/purposive non random sampling
JNRS = dataset[dataset['day']=='Sun']
print("Judgement Non Random Sample")
JNRS
```

### Emperical Rule

```python
# Emperical Rule
sns.distplot(sc)
```

- We can use the cdf() function of the scipy.stats.norm module to calculate the cumulative probability(area under a distribution curve).
- Syntax : cdf(x, mean, SD)
- Parameters :
   * x : value up to which cumulative probability is to be calculated
   * mean : mean of the distribution
   * SD : standard deviation of the distribution


```python
from scipy.stats import norm
```

```python
# setting the values of mean and S.D.
mean = sc.mean()
SD = sc.std()
  
# value of cdf between one, two and three S.D. around the mean
one_sd = norm.cdf(SD, mean, SD) - norm.cdf(-SD, mean, SD)
two_sd = norm.cdf(2 * SD, mean, SD) - norm.cdf(-2 * SD, mean, SD)
three_sd = norm.cdf(3 * SD, mean, SD) - norm.cdf(-3 * SD, mean, SD)
  
# printing the value of fractions within each band
print("Fracton of values within one SD(68% data) =", one_sd)
print("Fracton of values within two SD(95% data) =", two_sd)
print("Fracton of values within three SD(99.7% data) =", three_sd)
```

### Euclidean, Minkouski, Manhanttan, Hamming Distance

```python
from scipy.spatial import distance
```

```python
a = (1, 2, 3)
b = (4, 5, 6)
```

```python
#Euclidean Distance : represents the shortest distance between two points.
print(distance.euclidean(a, b))
```

```python
#Manhattan Distance : It is the sum of absolute differences between points across all the dimensions.
print(distance.cityblock(a,b))
```

```python
#Minkowski Distance : It is the generalized form of Euclidean and Manhattan Distance.
print(distance.minkowski(a,b))
```

```python
#Hamming Distance : It measures the similarity between two strings of the same length.
#The Hamming Distance between two strings of the same length is the number of positions at which the corresponding characters are different
string_1 = 'data' 
string_2 = 'science' 
str1 = 'euclidean'
str2 = 'manhattan'
len(string_1), len(string_2)
len(str1),len(str2)
hamming_distance = distance.hamming(list(str1), list(str2))
hamming_distance
```
